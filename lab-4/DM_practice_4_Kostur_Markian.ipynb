#### **Система Лоренца:**

Система Лоренца описується трьома диференціальними рівняннями:

$$
\begin{cases}
\frac{dx}{dt} = \sigma (y - x) \\
\frac{dy}{dt} = x (\rho - z) - y \\
\frac{dz}{dt} = xy - \beta z
\end{cases}
$$

де:
- $( \sigma )$ — коефіцієнт Прандтля (зазвичай $ \sigma = 10 $),
- $( \rho )$ — числовий параметр для турбулентності (зазвичай $ \rho = 28 $),
- $( \beta )$ — коефіцієнт відношення висоти до довжини (зазвичай $ \beta = 8/3 $).

#### **Об'єкти Нарендра-Пархтізаратхі:**

Об'єкти Нарендра-Пархтізаратхі описуються наступними рівняннями:

$$
\begin{cases}
x_1(t+1) = \alpha_1 x_1(t) + \beta_1 x_2(t) + \gamma_1 x_3(t) + \delta_1 \\
x_2(t+1) = \alpha_2 x_1(t) + \beta_2 x_2(t) + \gamma_2 x_3(t) + \delta_2 \\
x_3(t+1) = \alpha_3 x_1(t) + \beta_3 x_2(t) + \gamma_3 x_3(t) + \delta_3
\end{cases}
$$

де $(\alpha_i)$, $(\beta_i)$, $(\gamma_i)$ і $(\delta_i)$ — коефіцієнти для кожного з рівнянь.

### **Нелінійних рівнянь Нарендра-Партхізаратхі:**

### Параметри:
- $( a_1, a_2 )$, $( b_1, b_2 )$ — це коефіцієнти, які можуть бути варійовані для дослідження динаміки системи та впливу на прогнозування.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.linear_model import Ridge
from scipy.integrate import odeint
```

```python
def lorenz_system(state, t, sigma, rho, beta):
  x, y, z = state
  dxdt = sigma * (y - x)
  dydt = x * (rho - z) - y
  dzdt = x * y - beta * z
  return [dxdt, dydt, dzdt]

def narendra_parthasarathy_system(x, y, z, w, alpha_1, alpha_2, beta_1, beta_2, gamma_1, gamma_2, delta_1, delta_2):
    dx = alpha_1 * y - beta_1 * x * z
    dy = alpha_2 * x * z - beta_2 * y
    dz = gamma_1 * x - delta_1 * y * z
    dw = gamma_2 * y * z - delta_2 * w
    return dx, dy, dz, dw

def system(state, z, a1, a2, b1, b2):
    x, y = state
    dx_dt = -a1 * x + b1 * y**2/np.cos(x)
    dy_dt = a2 / np.sin(x) - (b2 * x * y)
    return [dx_dt, dy_dt]

def grid_search_model(model, param_grid, X_train, y_train):
    grid_search = GridSearchCV(model, param_grid, scoring='neg_mean_absolute_error', cv=3, n_jobs=-1)
    grid_search.fit(X_train, y_train)
    print(f"\nbest param {model.__class__.__name__}: {grid_search.best_params_}")
    return grid_search.best_estimator_

def train_and_evaluate_model(model, X_train, X_val, y_train, y_val):
    model.fit(X_train, y_train)
    y_val_pred = model.predict(X_val)
    mse = mean_absolute_error(y_val, y_val_pred)
    print(f"Mean Squared Error {model.__class__.__name__}: {mse:.4f}")

def train_and_evaluate_model_by_MSE_R2(model, X_train, X_val, y_train, y_val, X_test, y_test):
    model.fit(X_train, y_train)
    y_val_pred = model.predict(X_val)
    val_mse = mean_absolute_error(y_val, y_val_pred)
    val_r2 = r2_score(y_val, y_val_pred)
    print(f"{model.__class__.__name__} - Validation MSE: {val_mse:.4f}, R²: {val_r2:.4f}")
    y_test_pred = model.predict(X_test)
    test_mse = mean_absolute_error(y_test, y_test_pred)
    test_r2 = r2_score(y_test, y_test_pred)
    print(f"{model.__class__.__name__} - Test MSE: {test_mse:.4f}, R²: {test_r2:.4f}")

def compare_models(models, X_train, X_val, y_train, y_val, X_test, y_test):
    results = []
    for model in models:
        model.fit(X_train, y_train)
        y_val_pred = model.predict(X_val)
        val_mse = mean_absolute_error(y_val, y_val_pred)
        val_r2 = r2_score(y_val, y_val_pred)
        y_test_pred = model.predict(X_test)
        test_mse = mean_absolute_error(y_test, y_test_pred)
        test_r2 = r2_score(y_test, y_test_pred)
        results.append({'Model': model.__class__.__name__, 'Validation MSE': val_mse, 'Validation R²': val_r2, 'Test MSE': test_mse, 'Test R²': test_r2})
    
    return pd.DataFrame(results)
```

```python
start_point = 0
end_point = 100
count_of_objects = 10000

sigma = 10
rho = 28
beta = 8/3

initial_state = [1, 1, 1]

t = np.linspace(start_point, end_point, count_of_objects)
states = odeint(lorenz_system, initial_state, t, args=(sigma, rho, beta))
df_lorenz = pd.DataFrame(states, columns=['x', 'y', 'z'])
df_lorenz.to_csv('./csv/lorenz_dataset.csv', index=False)
```

```python
def plot_all_predictions(models_dict, X_tests, y_tests, system_names):
    """
    models_dict: a dictionary where the keys are system names and the values are lists of models for each system
    X_tests: a dictionary of test data for each system
    y_tests: a dictionary of actual values for each system
    system_names: a list of system names
    """
    fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(20, 20))
    axes = axes.flatten()
    plot_idx = 0

    for system_name in system_names:
        models = models_dict[system_name]
        X_test = X_tests[system_name]
        y_test = y_tests[system_name]
        
        for model in models:
            y_pred = model.predict(X_test)
            ax = axes[plot_idx]
            ax.plot(y_test, label='Actual', color='black', linestyle='dashed', linewidth=2)
            ax.plot(y_pred, label=f'{model.__class__.__name__} Predicted', alpha=0.7)
            ax.set_title(f'{system_name}: {model.__class__.__name__}')
            ax.set_xlabel('Sample Index')
            ax.set_ylabel('Value')
            ax.legend()
            plot_idx += 1

    plt.tight_layout()
    plt.show()

models_dict = {
    'Lorenz system': models,
    'Narendra-Parthasarathy system': models,
    'Custom system': models
}

X_tests = {
    'Lorenz system': X_test_lorenz_scaled,
    'Narendra-Parthasarathy system': X_test_narendra_scaled,
    'Custom system': X_test_custom_scaled
}

y_tests = {
    'Lorenz system': y_test_lorenz,
    'Narendra-Parthasarathy system': y_test_narendra,
    'Custom system': y_test_custom
}

system_names = ['Lorenz system', 'Narendra-Parthasarathy system', 'Custom system']

plot_all_predictions(models_dict, X_tests, y_tests, system_names)

```
